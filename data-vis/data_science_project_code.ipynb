## 5) Data Cleaning 

### b) Cleaning 2
*By \<Gabby>*

```python
data = pd.read_csv('jobs_in_data.csv')

# remove redundant cols, only domestic companies
data = data.drop(columns=['salary', 'salary_currency']).loc[(data['company_location']=='United States'),:]

# isolate entry level, full time obs
entry_data = data.loc[(data['experience_level']=='Entry-level') & (data['employment_type']=='Full-time'),:]

# drop unecessary cols
entry_data = entry_data.drop(columns = ['experience_level', 'employment_type', 'work_year', 'employee_residence', 'job_title'])
entry_data.head()
```

## 6) Data Analysis

### b) Analysis 2
*By \<Gabby>*

```python
small_data = entry_data.loc[entry_data['company_size']=='S',:]
medium_data = entry_data.loc[entry_data['company_size']=='M',:]
large_data = entry_data.loc[entry_data['company_size']=='L',:]

a = small_data[['salary_in_usd']].describe()
b = medium_data[['salary_in_usd']].describe()
c = large_data[['salary_in_usd']].describe()


sum_stats = pd.concat([a,b,c], axis=1)
sum_stats.columns = ['small', 'medium','large']

sum_stats
```

```python
# create salary ranges
bins = [0, 50000, 80000, 110000, 140000, 170000, 200000, 230000,300000]
labels = ['<50K', '50K-80K', '80K-110K', '110K-140K', '140K-170K', '170K-200K', '200K-230K', '>230K']
entry_data['salary_range'] = pd.cut(entry_data['salary_in_usd'], bins=bins, labels=labels)


fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# calc percentages by salary range within each company size
stacked_data = entry_data.groupby(['company_size', 'salary_range'], observed=False)['salary_in_usd'].count().unstack()
stacked_data_normalized = (stacked_data.T / stacked_data.sum(axis=1)).T  # Normalize to percentages

# plot stacked bar plot
stacked_data.plot(kind='bar', stacked=True, figsize=(12, 6), ax = axes[0],colormap='viridis')
axes[0].set_title("Count of Salary Ranges by Company Size")
axes[0].set_xlabel("Company Size")
axes[0].set_ylabel("Count")
axes[0].legend(title="Salary Range", bbox_to_anchor=(1.05, 1), loc='upper left')


stacked_data_normalized.plot(kind='bar', stacked=True, figsize=(12, 6), ax=axes[1], colormap='viridis')
axes[1].set_title("Percentage of Salary Ranges by Company Size")
axes[1].set_xlabel("Company Size")
axes[1].set_ylabel("Percentage")
axes[1].legend(title="Salary Range", bbox_to_anchor=(1.05, 1), loc='upper left')


plt.tight_layout()
plt.show()
```

```python
# calc percentages by salary range within each job category
stacked_data = entry_data.groupby(['job_category', 'salary_range'], observed=False)['salary_in_usd'].count().unstack()
stacked_data_normalized = (stacked_data.T / stacked_data.sum(axis=1)).T  # Normalize to percentages


stacked_data_normalized.plot(kind='bar', stacked=True, figsize=(12, 6),  colormap='viridis')
plt.title("Percentage of Salary Ranges by Job Category")
plt.xlabel("Job Category")
plt.ylabel("Percentage")
plt.legend(title="Salary Range", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xticks(rotation = 45, ha = "right")
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data=entry_data, x='job_category', y='salary_in_usd', color='pink')
plt.xticks(rotation=90)
plt.title('Salary Distribution Across Job Categories')
plt.xlabel('Job Category')
plt.xticks(rotation = 45, ha = "right")
plt.ylabel('Salary in USD') 

entry_data.groupby(['job_category'])[['salary_in_usd']].describe()
```

```python
entry_data['company_size_work_setting'] = entry_data['company_size'] + " & " + entry_data['work_setting']

category_size_work_setting = pd.crosstab(entry_data['job_category'], entry_data['company_size_work_setting'])

# plot the stacked bar chart
category_size_work_setting.plot.bar(stacked=True, colormap='Set3', figsize=(12, 6))
plt.xticks(rotation=90)
plt.title('Job Category Distribution by Company Size and Work Setting')
plt.xlabel('Job Category')
plt.xticks(rotation = 45, ha = "right")
plt.ylabel('Count')
plt.legend(title='Company Size & Work Setting', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
